```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Frontline_Occupation_Demographics (3).csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df['demograph_total'] =  df['non-hispanic white']+ df['non-hispanic black'] + df['non-hispanic american indian'] + df['non-hispanic asian'] + df['non-hispanic pacific islander'] + df['non-hispanic other'] + df['non-hispanic two or more races'] + df['hispanic']

df.head()
```

```python
df['PCT-non-hispanic white'] = round(df['non-hispanic white']/ df['demograph_total'] *100, 0)
df['PCT-non-hispanic black'] = round(df['non-hispanic black']/ df['demograph_total'] *100, 0)
df['PCT-hispanic'] = round(df['hispanic']/ df['demograph_total'] *100, 0)
df['PCT-non-hispanic asian'] = round(df['non-hispanic asian']/ df['demograph_total'] *100, 0)
df
```

```python
df.columns
```

```python
ax = df[['PCT-non-hispanic white',
       'PCT-non-hispanic black', 'PCT-hispanic', 'PCT-non-hispanic asian']].plot.bar(figsize = (12,10))
plt.ylabel('Percentage of Workers')
plt.title('Essential Workers by Race')
plt.xlabel('Essential Jobs')
ax.set_xticklabels(df['Occupation'])
```

```python
ax = df['Avg Occupants/Bedrooms Ratio'].plot.bar()
plt.ylabel('Occupants per bedroom')
plt.title('Occupants per bedroom for essential workers')
plt.xlabel('Essential Jobs')
ax.set_xticklabels(df['Occupation'])
# use axhline
mean = df['Avg Occupants/Bedrooms Ratio'].mean()
ax.axhline(mean)
plt.annotate(
    "mean", xy=(0, df['Avg Occupants/Bedrooms Ratio'].mean() + .02))
def add_value_labels(ax, spacing=5):
    # For each bar: Place a label
    for rect in ax.patches:
        # Get X and Y placement of label from rect.
        y_value = rect.get_height()
        x_value = rect.get_x() + rect.get_width() / 2

        # Number of points between bar and label. Change to your liking.
        space = spacing
        # Vertical alignment for positive values
        va = 'bottom'

        # If value of bar is negative: Place label below bar
        if y_value < 0:
            # Invert space to place label below
            space *= -1
            # Vertically align label at top
            va = 'top'

        # Use Y value as label and format number with one decimal place
        label = "{:,.2f}".format(y_value)

        # Create annotation
        ax.annotate(
            label,                      # Use `label` as label
            (x_value, y_value),         # Place label at end of the bar
            xytext=(0, space),          # Vertically shift label by `space`
            textcoords="offset points", # Interpret `xytext` as offset in points
            ha='center',                # Horizontally center label
            va=va)                      # Vertically align label differently for
                                        # positive and negative values.


# Call the function above. All the magic happens there.
add_value_labels(ax)
```
